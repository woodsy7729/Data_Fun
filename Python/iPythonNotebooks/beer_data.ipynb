The following is me just practicing with Data Analytics. I play around with organising and grouping the data, as well as visualisation with the Bokeh package.

```python
from bokeh.io import output_notebook, show
output_notebook()
```

```python
import pandas as pd
path = "../Dependencies/strata_yt/strata_data/data/beer2.csv"
beer = pd.read_csv(path)
taste_style = beer[beer['review_taste']==5].groupby('beer_style')['beer_id'].count().sort_values(ascending=False)
taste_style=taste_style.reset_index()
```

```python
from bokeh.plotting import figure
from bokeh.models import ColumnDataSource
source = ColumnDataSource(taste_style)
dfList = taste_style['beer_style'].tolist()

p = figure(x_range=dfList, plot_height=650, title="Beer Style Counts w/ 5/5 taste")
p.vbar(x='beer_style', top='beer_id', width=0.6,source=source)
p.circle(x='beer_style', y='beer_id',source=source,size=10, line_color="red", fill_color="white")
p.line(x='beer_style', y='beer_id',source=source, color="red", line_width=1)
p.xaxis.major_label_orientation = 20
p.y_range.range_padding = 0
show(p)
```

```python
most_of_style = beer.groupby('beer_style').count()['beer_name'].sort_values(ascending=False).reset_index().head(10)
dfList = most_of_style['beer_style'].tolist()
data = ColumnDataSource(most_of_style)
plot= figure(x_range=dfList, title="Most of Style Counts")
plot.vbar(x="beer_style", top='beer_name', width=0.6, source=data)
plot.xaxis.major_label_orientation = 20
plot.y_range.start =0
show(plot)
```

```python
test_data = beer

test_data['mean'] = (test_data['review_appearance'] + test_data['review_taste'] + test_data['review_aroma'] + test_data['review_overall'] + test_data['review_palate'])/5
counts=test_data.groupby('beer_style').count()
counts = test_data.groupby('beer_style').count()['mean'].tolist()
means = test_data[['beer_style', 'mean']].groupby('beer_style').mean()
means['count'] = counts
top_mean = means.sort_values(by='mean', ascending=False).head(10)
top_count = means.sort_values(by='count', ascending=False).head(10)
means= means.reset_index()
```

```python
from bokeh.models import LinearAxis, Range1d
from bokeh.transform import dodge
meansdata = ColumnDataSource(means)
dflist = means['beer_style'].tolist()
meansplot = figure(title='Count and Means', x_range=dflist)
meansplot.xaxis.major_label_orientation = 20
meansplot.extra_y_ranges = {"foo": Range1d(start=0, end=5)}
meansplot.yaxis.axis_label = "Count"
meansplot.add_layout(LinearAxis(y_range_name="foo", axis_label='Average Rating'), 'right')
meansplot.vbar(x=dodge('beer_style', 0.2, range=meansplot.x_range), top='mean',color="green", y_range_name="foo", width=0.2, source =meansdata )
meansplot.vbar(x='beer_style', top ='count', width=0.2,source=meansdata, color="red" )

meansplot.legend.location = "top_center"
meansplot.y_range.start = 0
meansplot.xaxis.axis_label = "Beer Style"
show(meansplot)
```
